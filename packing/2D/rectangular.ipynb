```python
import matplotlib.pyplot as plt
import numpy as np
num=100
x = np.linspace(0,4,num=num)
y = np.linspace(0,1,num=num)

def f(x, y):
    return y * np.sin(x) 

X, Y = np.meshgrid(x,y)
Z = np.zeros((num,num))

for i in range(num):
    for j in range(num):
        Z[i,j] = f(X[i,j],Y[i,j])
plt.pcolor(X, Y, Z)
plt.show()
```

```python
def create_random_porous_media():
    nx, ny = (50, 20)
    x = np.linspace(0, 10, nx)
    y = np.linspace(0, 10, ny)
    xv, yv = np.meshgrid(x, y)
    
    print(yv.shape)
    plt.figure(figsize=(10, 10))
    
    fig = plt.gcf()
    for j in range(xv.shape[0]):
        for i in range(yv.shape[0]):
            circle1 = plt.Circle((xv[i,j],yv[i,j]), 0.1, color='r')    
        fig.gca().add_artist(circle1)
    plt.xlim(0, 10)
    plt.ylim(0, 10)
#     plt.savefig('test.png')
    return xv,yv
xv,yv = create_random_porous_media()
```
