```python
from scipy.stats import linregress
```

```python
label = 'C'
g1,g2,g3 = df.iloc[0:6],df.iloc[6:12],df.iloc[12:]
points = []
slopes = []
llim1,llim2,llim3 = 0., 0., 0. 
hlim1,hlim2,hlim3 = 5, 5. , 5. 

data1 = g1[['t',label]]
d1 = data1[(data1[label]!=0) & (data1[label]>=llim1) & (data1[label]<hlim1)]
points.append(list(d1.index))
x = d1.t.values
y = d1[label].values
plt.scatter(x,y)
plt.ylim(0,max(y)+max(y)/2)
s = linregress(x,y)
plt.plot(x, s[0] * np.array(x) + s[1], color='black')
print 'slope: ', s[0],'intercept: ', s[1], 'R^2', s[2]**2
slopes.append([s[0],s[1],s[2]**2])
data2 = g2[['t',label]]

d2 = data2[(data2[label]!=0) &(data2[label]>=llim2) & (data2[label]<hlim2)]
points.append(list(d2.index))

x = d2.t.values
y = d2[label].values
 
plt.scatter(x,y)
plt.ylim(0,max(y)+max(y)/2)

s = linregress(x,y)
plt.plot(x, s[0] * np.array(x) + s[1], color='black')

print 'slope: ', s[0],'intercept: ', s[1], 'R^2', s[2]**2
slopes.append([s[0],s[1],s[2]**2])

data3 = g3[['t',label]]

d3 = data3[(data3[label]!=0) &(data3[label]>=llim3) & (data3[label]<hlim3)]

points.append(list(d3.index))

x = d3.t.values
y = d3[label].values

plt.scatter(x,y)
plt.ylim(0,max(y)+max(y)/2)

s = linregress(x,y)
plt.plot(x, s[0] * np.array(x) + s[1], color='black')

print 'slope: ', s[0],'intercept: ', s[1], 'R^2', s[2]**2
slopes.append([s[0],s[1],s[2]**2])
```
